```python
import pandas as pd
import numpy as np
```

```python
def load_data():
    data = pd.read_excel('untuk_testing.xlsx')
    return data
```

```python
df = load_data()
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
nilai = [analyser.polarity_scores(x) for x in df['Cleaned_Text']]
print(nilai)
df['Compound_Score'] = [x['compound'] for x in nilai]
```

```python
df.head()
```

```python
df.nsmallest(1000, ['Compound_Score'])
```

```python
df.loc[df['Compound_Score'] == 0, 'Sentimen'] = 'negatif'

df.loc[df['Compound_Score'] > 0, 'Sentimen'] = 'positif'

df.nsmallest(100, ['Compound_Score'])
```

```python
df.to_excel("data_training_lexicon.xlsx")
```
