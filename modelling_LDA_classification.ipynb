```python
tknzr = TweetTokenizer()

def tokenize(text):
    lda_tokens = []
    tokens = tknzr.tokenize(text)
    for token in tokens:
        lda_tokens.append(token)
    return lda_tokens

def prepare_text_for_lda(text):
    tokens = tokenize(text)
    return tokens
```

```python
from gensim import corpora
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel

#import pyLDAvis.gensim
import gensim
```

```python
dictionary = gensim.corpora.Dictionary.load('dictionary_lda.gensim')
corpus = pickle.load(open('corpus_lda.pkl', 'rb'))
ldamodel = gensim.models.ldamodel.LdaModel.load('modelLDA_5.gensim')

text_data = []
for line in df["caption"]:
    tokens = prepare_text_for_lda(line)
    new_doc_bow = dictionary.doc2bow(tokens)
    results=ldamodel.get_document_topics(new_doc_bow)
    text_data.append({'tweet':line, 'topic':results})
```

```python
text_data
```

```python
data=pd.DataFrame(text_data)
data
```

```python
data_tweet = data.tweet
data_topic = data.topic
```

```python
result_tweet=[]
for i in range(len(data_topic)):
    result_topic=max(data_topic[i], key=lambda x: x[1])
    result_tweet.append({'tweet':data_tweet[i], 'topic':result_topic[0], 'score':result_topic[1]})
```

```python
result_topic=pd.DataFrame(result_tweet)
result_topic
```
